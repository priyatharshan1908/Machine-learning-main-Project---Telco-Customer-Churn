```python
# Impute a null values using (mode) or (median)
# If columns are numerical using "mean"
data1["tenure"].fillna(data1["tenure"].mean,inplace=True)
data1["MonthlyCharges"].fillna(data1['MonthlyCharges'].mean,inplace=True)
data1["TotalCharges"].fillna(data1["TotalCharges"].mean,inplace = True)
# If colomns are categorical using "mode"
data1["InternetService"].fillna(data1["InternetService"].mode()[0],inplace=True)
data1["TechSupport"].fillna(data1["TechSupport"].mode()[0],inplace = True)
data1["Contract"].fillna(data1["Contract"].mode()[0],inplace=True)
data1["PaymentMethod"].fillna(data1["PaymentMethod"].mode()[0],inplace =True)
```
